```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

np.random.seed(42)

data_root = '../datasets/lifesat/'
df = pd.read_csv( data_root + 'lifesat.csv' )

df.rename( columns={'GDP per capita (USD)':'gdp', 'Life satisfaction':'lifesat' }, inplace=True )
df.plot( x='gdp', y = 'lifesat', kind='scatter' )

X = df[['gdp']].to_numpy()
y = df[['lifesat']].to_numpy()

x_values = np.arange( min( df['gdp'] ), max( df['gdp'] ), 100 ).reshape( -1, 1 )

model = LinearRegression( )
model.fit(X, y)
y_values = model.predict( x_values )
plt.plot( x_values, y_values )

model = KNeighborsRegressor( n_neighbors=3)
model.fit(X, y)
y_values = model.predict( x_values )
plt.plot( x_values, y_values )
```
